Some ideas for the tea maker:

•	it would know the tea brand and set the water temperature accordingly
•	tea cups would have water level sensors, with custom setpoint per cup, so the tea maker would know when to start boiling more water
•	in a work place, there would be a connection to the calendar, and it would know when the employees have a break
•	it would also know from the monitoring system of working hours that if someone is doing overtime, it would make tea for the employee
•	it would learn your tea drinking habits and make custom profiles for the tea drinkers (brand, amount, frequency) and it would regulate itself accordingly
•	it would know how to search information about different tea brands from the internet
•	computer vision that would read the information from the tea packages
•	facial recognition to identify the user
•	pressure sensor on the bench to recognize that a user is present


Analyze of the tea maker as a system
•	OP Peculiar operation: boils water and makes tea with intelligent monitoring system that regulates it's operation with possibility to find the needed information
•	C1 Conditions of production: boiler, water, electricity, measurement units, tea, the tea drinkers, a database o「 a internet connection, a smart controller
•	C2 Conditions of reproduction: proper maintenance of the tea maker, connection to the power grid, running water, a tea distributor, service personnel
•	C3 External conditions: electricity production, a disesase which destroys all tea plants, training facility of the tea maker maintainer

Rationalities

•	Classical/positivist: Amount of tea drinkers that use the tea maker during some period of time can be calculated.
•	Non-classical/non-positivist: The willingness of people to use the tea maker can be evaluated and modelled.
•	Post-non-classical/critical: Before holidays people do more overtime at work, so they drink more tea to stay focused. In such situation the tea maker would make tea also during worktime and not just during brakes. Also automatically serve the tea to the users tea cup.
•	One possible intervention to modify the system dynamics is to serve too much tea during time before holidays which would lead to unwillingness to drink tea due to getting sick of excess tea.


Signal path in the tea maker
Signal could travel as follows:

1.	signal from monitoring system of working hours (break, overtime or time before holidays)
2.	pressure indication from the bench to the tea maker: switch the tea maker on/off
3.	tea maker on-signal leads to start facial recognition from the tea maker to the cup
4.	visual data from the cup's camera to the tea maker's facial recognition system
5.	user's profile selected by the facial recognition system from the database
6.	signal to select the correct tea brand
7.	boil the water to the correct temperature and make tea
8.	signal to serve the tea to the cup
9.	signal to start monitoring tea level in the cup from the tea maker
10.	signal from the cup to the tea maker that the tea is running out

Node O has integrated calendar and clock (or it could get the info from the Internet), and it sends automatically the information to node 1 about whether it is proper time to make tea or not. The change in pressure sensor status tells the node 1 if the drinker is present. Node 1 combines pressure sensor information with information from node O and sets the tea maker on or off. When the on/off status changes, the change is sent to nodes 2 and 6.

```python
import networkx as nx
import matplotlib.pyplot as plt
# hide warnings as matplotlib might generate a bunch 
import warnings
warnings.filterwarnings ('ignore')
print("networkx  version:  "+nx.__version__) 
G = nx.DiGraph()
G.add_node(0)
for x in range(8): 
    G.add_edge(x+1,x+2)
G.add_edge(9,6) 
G.add_edge(0,1) 
G.add_edge(1,6)

# draw the graph plt.subplot(121)
nx.draw_circular(G, with_labels=True, font_weight='bold')
```

Analysis of the topology as a communication network (i.e. how data travels in the network)
If the topology would be a complete ring topology from node O to 8, it would not be very efficient because signal would have to travel through nodes which are not related to the signal in question. This would also execute some unnecessary functions which would slow down the tea making process and waste energy.

The signal (or data) between the links is different, so the speed of link can vary. The links are uni-directional: the previous node does not need direct feedback 「f om the next node, because the signals are processed in a chain. Cluster coefficient is zero, so there are no shortcuts. Data must go through all nodes in that path.

```python
# map nodes 0-9 to A-J 
mapping={0:'A',1:'B',2:'C',3:'D',4:'E',5:'F',6:'G',7:'H',8:'I',9:'J'}
G    =    nx.relabel_nodes(G,mapping)
# draw the network
nx.draw(G, with_labels=True, font_weight='bold') 
plt.axis('off')
plt.show()
```


The tea maker has the physical process T, which is sensed by nodes A, B, D, G and J.

Nodes C, E, F, H, and I only see T through the sensing nodes.

The following nodes aqcuire and process data of supporting reality:

A.	signal from monitoring system of working hours (break, overtime or time before holidays, current time) (reality: time)

B.	pressure indication from the bench to the tea maker: switch the tea maker on/off (reality: pressure)

D. visual data from the cup's camera to the tea maker's facial recognition system (reality: image)

G. boil the water to the correct temperature and make tea (reality: temperature)

J. signal from the cup to the tea maker that the tea is running out (reality: level of liquid)

The following nodes only receive data from supporting reality

C. tea maker on-signal leads to start facial recognition from the tea maker to the cup

E. user's profile selected by the facial recognition system from the database

F. signal to select the co「rect tea brand

H. signal to serve the tea to the cup

I. signal to start monitoring tea level in the cup from the tea maker
