**(4) Consider a house with an electric heater. How the electricity consumption of the heater would be affected by the outside air temperature? Discuss about 3 current heating methods and their differences.**

Answer:
The electricity consumption of an electric heater in a house would be affected by the outside air temperature in a few ways. First, if the outside temperature is very cold, the heater may need to work harder to maintain the desired inside temperature, resulting in higher electricity consumption. Second, if the outside temperature is milder, the heater may not need to work as hard and therefore consume less electricity.

Three current heating methods and their differences are:

Electric resistance heating: This method uses electric coils or elements to generate heat. It is one of the simplest and most common types of electric heating, but it is also the least efficient.

Electric heat pumps: This method uses electricity to move heat from one location to another. It is more efficient than electric resistance heating because it takes advantage of the heat in the air or ground outside to warm the house, instead of generating heat directly.

Electric radiant heating: This method uses electric cables or mats placed under the floor or embedded in walls to radiate heat. It is a very efficient method of heating because it heats objects directly, instead of heating the air.

Overall, Electric heat pumps are more efficient than Electric resistance heating and Electric radiant heating is most efficient among these three heating methods.
